#  Memory Management in LangChain

**Buffer Memory**:

---

### **ConversationBufferMemory (LangChain)**

* Keeps a **list of previous chat messages**.
* Passes the **entire chat history** into the prompt template.
* Useful for **context-aware conversations**.
* **Does not summarize**; keeps full history (can grow large).

**Key Points:**

* Stores messages in a buffer.
* Helps LLM remember past conversation.
* Simple, raw memory (no compression).

---



---

### ✅ Key Takeaways

* **ConversationBufferMemory** keeps a running list of **all chat messages**.
* Every new input sees **all previous messages** so replies are consistent and context-aware.
* No summarization — it’s a **raw buffer** of conversation history.
