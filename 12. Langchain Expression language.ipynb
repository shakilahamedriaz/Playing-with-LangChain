# 📝 LCEL (LangChain Expression Language) Agenda

### 1. **Definition**

* LCEL হলো একটি declarative language (DSL) → সহজে LLM pipeline/chain define করার জন্য।
* Functional programming style, readable + composable।

👉 *LCEL মানে: LLM flow বানানোর জন্য শর্টকাট ভাষা।*

---

### 2. **Core Goals**

* **Composable** → ছোট কম্পোনেন্ট দিয়ে বড় চেইন বানানো।
* **Readable** → কোড কম লিখে সহজে বোঝা।
* **Flexible** → বিভিন্ন LLM, prompt, retriever, memory connect করা যায়।
* **Efficient** → optimized execution (streaming, async ইত্যাদি) support করে।

---

### 3. **Main Features**

1. **Operators**

   * `|` (pipe) → একটার output → পরের input
   * `+` (combine) → একাধিক আউটপুট merge করা
   * `map / bind` ইত্যাদি functional concepts

2. **Components**

   * PromptTemplate
   * LLM / ChatModel
   * OutputParser
   * Retriever / Tool
   * Memory

3. **Execution Modes**

   * Sync
   * Async
   * Streaming

---



### Example LCEL Pipeline

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_template("Translate to French: {text}")
model = ChatOpenAI()
parser = StrOutputParser()

# LCEL chain
chain = prompt | model | parser

result = chain.invoke({"text": "Hello, how are you?"})
print(result)  # "Bonjour, comment ça va?"
```
