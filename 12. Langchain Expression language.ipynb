#  LCEL (LangChain Expression Language) Agenda

### 1. **Definition**

* LCEL   declarative language (DSL)   LLM pipeline/chain define  
* Functional programming style, readable + composable

 *LCEL : LLM flow    *

---

### 2. **Core Goals**

* **Composable**       
* **Readable**      
* **Flexible**   LLM, prompt, retriever, memory connect  
* **Efficient**  optimized execution (streaming, async ) support 

---

### 3. **Main Features**

1. **Operators**

   * `|` (pipe)   output   input
   * `+` (combine)    merge 
   * `map / bind`  functional concepts

2. **Components**

   * PromptTemplate
   * LLM / ChatModel
   * OutputParser
   * Retriever / Tool
   * Memory

3. **Execution Modes**

   * Sync
   * Async
   * Streaming

---



### Example LCEL Pipeline

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_template("Translate to French: {text}")
model = ChatOpenAI()
parser = StrOutputParser()

# LCEL chain
chain = prompt | model | parser

result = chain.invoke({"text": "Hello, how are you?"})
print(result)  # "Bonjour, comment a va?"
```
